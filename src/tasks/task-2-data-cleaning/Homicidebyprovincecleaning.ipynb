```python
# Read the excel file
import pandas as pd
import numpy as np
import os
import glob
```

```python
def read_excel_csv_files(directory_path):
    # Change the working directory to the desired path
    os.chdir(directory_path)

    # Get a list of all Excel, XLSX, and CSV files in the directory
    file_list = glob.glob("*.xls") + glob.glob("*.xlsx") + glob.glob("*.csv")

    # Initialize an empty dictionary to store DataFrames
    dataframes = {}

    # Loop through the files and read each one into a pandas DataFrame
    for file in file_list:
        file_key = os.path.splitext(file)[0]  # Remove the file extension
        if file.endswith('.csv'):
            df = pd.read_csv(file, )
        elif file.endswith(('.xls', '.xlsx')):
            df = pd.read_excel(file)
        
        # Add the DataFrame to the dictionary with the filename as the key
        dataframes[file_key] = df

    return dataframes

# Example usage of the function:
directory_path = r'C:\Users\chimi\Desktop\Python Data Science Projects\Italy Crime\Data\Crime Rate data\Regional Data'
dataframes = read_excel_csv_files(directory_path)

# Print the keys in the 'dataframes' dictionary
print("Keys in the 'dataframes' dictionary:")
for key in dataframes.keys():
    print(key)

# Access and assign the DataFrames using their keys (filenames)
# Replace 'filename1', 'filename2', and 'filename3' with the actual keys
# Access and assign the DataFrames using their keys (filenames)
# Replace 'filename1', 'filename2', and 'filename3' with the actual keys
df1 = dataframes['prov_df']
df2 = dataframes['region capital commited crime 17-21']
df3 = dataframes['type of murder - big mun - 2006 to 2021']

# Print the separate DataFrames
print("\nDataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
print("\nDataFrame 3:")
print(df3)


```
