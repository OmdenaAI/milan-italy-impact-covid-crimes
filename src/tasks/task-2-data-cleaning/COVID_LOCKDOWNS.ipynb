```python
import pandas as pd
```

```python
# set header to row 3
lockdowns = pd.read_csv('../../data/COVID LOCKDOWNS IN ITALY (MM-DD-YYYY FORMAT) - Sheet1.csv', header=4)
lockdowns
```

```python
# count NaN values
lockdowns.isna().sum()
```

```python
# show NaN rows in Code
lockdowns[lockdowns['Code'].isna()]
```

```python
# display unique values in Code
lockdowns['Code'].unique()
```

```python
# set Code to 'NA' for NaN rows
lockdowns.loc[lockdowns['Code'].isna(), 'Code'] = 'NA'
```

```python
# check if is there NaN values in Code
lockdowns[lockdowns['Code'].isna()]
```

```python
# show unique combination of Lockdown Start and Lockdown End
lockdowns[['Lockdown Start', 'Lockdown End']].drop_duplicates()
```

```python
# combine Lockdown Start and Lockdown End
lockdowns['Lockdown'] = lockdowns['Lockdown Start'] + ' to ' + lockdowns['Lockdown End']
lockdowns['Lockdown.1'] = lockdowns['Lockdown Start.1'] + ' to ' + lockdowns['Lockdown End.1']
lockdowns['Lockdown.2'] = lockdowns['Lockdown Start.2'] + ' to ' + lockdowns['Lockdown End.2']
lockdowns['Lockdown.3'] = lockdowns['Lockdown Start.3'] + ' to ' + lockdowns['Lockdown End.3']
lockdowns['Lockdown.4'] = lockdowns['Lockdown Start.4'] + ' to ' + lockdowns['Lockdown End.4']
lockdowns['Lockdown.5'] = lockdowns['Lockdown Start.5'] + ' to ' + lockdowns['Lockdown End.5']

# drop columns
lockdowns.drop(columns=['Lockdown Start', 'Lockdown End', 'Lockdown Start.1', 'Lockdown End.1', 'Lockdown Start.2', 'Lockdown End.2', 'Lockdown Start.3', 'Lockdown End.3', 'Lockdown Start.4', 'Lockdown End.4', 'Lockdown Start.5', 'Lockdown End.5'], inplace=True)

lockdowns
```

```python
lockdowns.isna().sum()
```

```python
# fill NaN values with 'did not lockdown'
lockdowns.fillna('did not lockdown', inplace=True)
```

```python
# save to csv
lockdowns.to_csv('../../data/COVID LOCKDOWNS IN ITALY_clean.csv', index=False)
```
